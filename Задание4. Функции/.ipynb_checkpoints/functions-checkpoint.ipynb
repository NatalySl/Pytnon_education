Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def find_owner(doc_number, data):
    for doc in data:
        if doc_number == doc['number']:
            return print(doc['name'])
    return print('Документ не найден')
    
def find_shelf(doc_number, direct):
    for key, value in direct.items():
        if doc_number in value:
            return print(key)
    return print('Документ не найден')

def List(doc):
    for el in doc:
        print(f' {el["type"]} "{el["number"]}" "{el["name"]}"')
                     
def add_shelf(shelf_number, direct):
    for key in direct.keys():
        if shelf_number == key:
            return print('полка уже существует')
    direct[shelf_number] = []
    return direct
              
def move(doc_number, shelf_number, direct):
    if shelf_number not in directories.keys():
        return print ('полка не найдена')
    for value in direct.values():
        for num in value:
            if doc_number in num:
                value.remove(doc_number)
                print('документ успешно перемещен ')
                return direct[shelf_number].append(doc_number)
    return print ('документ не найден')
              
def delete_doc (document_number,doc):
        for el in doc:
            if document_number == el['number']:
                return documents.remove(el)
        return print('документ не найден')

def delete_shelf (document_number,direct):
        for value in direct.values():
            for i in value:
                if i == document_number:
                    value.remove(i)
                    return print ('Документ успешно удален')
        return print('документ не найден')
              
def add(document_number, document_type, fn, shelf_number,doc,direct):
    for key in direct.keys():
        if shelf_number == key:
            doc.append ({'type': document_type, 'number': document_number, 'name': fn})
            direct[shelf_number].append(document_number)
            return print('документ успешно добавлен')
    return print('полка не существует')  

def main():
    while True:
        user_input = input('Введите команду: p – узнать кому принадлежит документ; s – узнать номер полки, на которой находится документ;l – вывести список всех документов; as – add shelf – добавить полку в перечень; d – удалить документ; m – переместить документ; a – добавить новый документ, q - прекратить работу.') 
        if user_input == 'p':
            doc_number = input('Введите номер документа ')
            find_owner(doc_number,documents)
              
        elif user_input == 's':
            doc_number = input('Введите номер документа ')
            find_shelf(doc_number,directories)
              
        elif user_input == 'l':
            List(documents)
              
        elif user_input == 'as':
            shelf_number = input('Введите номер полки ')
            add_shelf(shelf_number,directories)
              
        elif user_input == 'd':
            doc_number = input('Введите номер документа ')
            delete_doc (doc_number, documents)
            delete_shelf (doc_number,directories)
              
        elif user_input == 'm':
            doc_number = input('Введите номер документа ')
            shelf_number = input('Введите номер полки ')
            move(doc_number, shelf_number,directories)
              
        elif user_input == 'a':
            doc_number = input('Введите номер документа ')
            document_type = input('Введите тип документа ')
            fn = input('Введите фамилию и имя владельца ') 
            shelf_number = input('Введите номер полки ')
            add(doc_number,document_type,fn,shelf_number,documents,directories)
              
        elif user_input == 'q':
            break
main()
print(f'directories = {directories}')
print(f'documents = {documents}')             
```
