Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
data = pd.read_csv('power.csv')
data.head()
#'Latvia','Estonia', 'Lithuania'
```

```python
data_country = data[ (data['country'] == 'Latvia') | (data['country'] == 'Lithuania') | (data['country'] == 'Estonia')]
data_country.tail()
```

```python
data_categoty = data_country[ ((data_country['category'] == 4) | (data_country['category'] == 12) | (data_country['category'] == 22)) &  (data_country['quantity']> 0)]
data_categoty.head()
```

```python
data_final = data_categoty[(data_categoty['year']>=2005) & (data_categoty['year']<=2010) ]
data_final.head()
```

```python
answer = sum(data_final['quantity'])
print(answer)
```
